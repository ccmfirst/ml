```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
```

```python
def f(x):
    return x**2 + 5 * x + 24

def fd(x):
    return 2*x + 5

r = 0.1
e = 10**-10

def gd1():
    x1 = 25
    y1 = f(x1)
    iMax = 1000
    i = 0
    while i < iMax:
        i += 1
        x2 = x1 - fd(x1) * r
        y2 = f(x2)
        print('iteration {}\tx = {:.2f}\t y = {:.2f}'.format(i,x2,y2))
        if np.abs(x2-x1) < e or np.abs(y2-y1) < e:
            break

        x1 = x2
        y1 = y2
    return x2, y2

gd1()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import  MaxNLocator
from matplotlib import animation

def f(x, y):
    return -1 * np.sin(0.5*x**2-0.25*y**2+3) * np.cos(2*x+1-np.exp(y))

def fdx(x, y):
    v = 0
    v += 2 * np.sin(2 * x - np.exp(y) + 1) * np.sin(0.5 * x**2 - 0.25 * y**2 + 3)
    v += x * np.cos(2 * x - np.exp(y) + 1) * np.cos(0.5 * x**2 - 0.25 * y**2 + 3)
    return v

def fdy(x, y):
    v = 0
    v += -1 * np.exp(y) * np.sin(2 * x - np.exp(y) + 1) * np.sin(0.5 * x**2 - 0.25 * y**2 + 3)
    v += 0.5 * y * np.cos(2 * x - np.exp(y) + 1) * np.cos(0.5 * x**2 - 0.25 * y**2 + 3)
    return v

r = 0.5
e = 10**-5

def gd2():
    x1 = -0.3
    y1 = 0.4
    z1 = f(x1, y1)
    iMax = 1000
    i = 0
    while i < iMax:
        i += 1
        x2 = x1 - fdx(x1, y1) * r
        y2 = y1 - fdy(x1, y1) * r
        z2 = f(x2, y2)
        print('iteration {}\t x = {:.2f}\t y = {:.2f}\t z = {:.2f}'.format(i, x2, y2, z2))
        if np.abs(x2-x1) < e or np.abs(y2 - y1) < e or np.abs(z2 - z1) < e:
            break

        x1 = x2
        y1 = y2
        z1 = z2

    return x2, y2, z2

gd2()
```
